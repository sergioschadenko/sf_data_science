```python
import pandas as pd
```

```python
orders = pd.read_csv("data/orders.csv", sep=";")
```

```python
products = pd.read_csv("data/products.csv", sep=";")
```

```python
orders.head()
```

```python
products.head()
```

```python
new_data = orders.merge(
    products,
    left_on="ID товара",
    right_on="Product_ID",
    how="left"
)
```

```python
new_data.head()
```

```python
new_data["Итого"] = new_data["Количество"] * new_data["Price"]
new_data.info()
```

```python
mask = (new_data["Оплачен"] == "Да")
new_data[mask].groupby(by="ID Покупателя")["Итого"].sum()
```
